**Experiment 1:** Can you find an epsilon value that increases the total reward.

Update the RunExperiment function to retun the total rewad:     
`return summedreward[N-1],count,data,averagereward`


```python
highest_reward=0
highest_eps=0
counter = 0
eps = 0

while True:
    total_reward,count,data,averagereward = run_experiment(mu, sigma, N, eps)
    if total_reward > highest_reward:
        highest_reward = total_reward
        highest_eps = eps
    
    if counter % 1000 == 0:
        print("iteration: {}  current epsilon: {}  total_reward: {}".format(counter, round(eps,4),total_reward))
    
    eps += 0.0001
    counter +=1
    
    if eps >= 1:
        break
print("Higest Reward is: {} when Epsilon is: {}".format(highest_reward,round(highest_eps,4))) 

plot_highest(count,data,averagereward)
```

By brute force iteration, we were able to determine the Highest  Reward is 1105.5061310902825 when Epsilon is set to 0.012.
